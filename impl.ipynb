```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Drawing environment using map.txt
def draw_map(map_file):
    x_list_start = []
    y_list_start = []
    x_list_end = []
    y_list_end = []

    with open(map_file) as f:
        for line in f:
            x_list_start.append((line.split()[0]))
            y_list_start.append((line.split()[1]))
            x_list_end.append((line.split()[2]))
            y_list_end.append((line.split()[3]))

    return x_list_start, y_list_start, x_list_end, y_list_end

#Drawing path using path.txt
def draw_path(path_file):
    x_list = []
    y_list = []

    with open(path_file) as f:
        for line in f:
            x_list.append((line.split()[0]))
            y_list.append((line.split()[1]))

    return x_list, y_list
```

```python
# Plot in x,y planes
x_list_start, y_list_start, x_list_goal, y_list_goal = draw_map("map.txt")
plt.axis('equal')
plt.axis([0, 30, 0, 30])

for i in range(len(x_list_start)):
    plt.plot([x_list_start[i], y_list_start[i]], [x_list_goal[i], y_list_goal[i]], color='black', linewidth=2)

#Path plotting 
x_list, y_list = draw_path("path1.txt")
x_list_2, y_list_2 = draw_path("path2.txt")
plt.plot(x_list, y_list, color='red', linewidth=2)
plt.plot(x_list_2, y_list_2, color='blue', linewidth=2)
    
plt.show()
```
